```python
import numpy as np
import matplotlib.pyplot as plt           # using numpy and matplotlib library for image Manupulation
from PIL import Image, ImageOps
```

```python
img = np.array(Image.open(r"C:\Users\navna\Downloads\mountain.jpg"))         # for showing image 
plt.figure(figsize=(8,8))
plt.imshow(img)
plt.show()
```

```python
print('# of dims: ',img.ndim)     # dimension of an image
print('Img shape: ',img.shape)    # shape of an image
print('Dtype: ',img.dtype)        # type of image
print(img[20, 20])                # pixel value at [R, G, B]
print(img[:, :, 2].min())     
```

```python
path = r"C:\Users\navna\Downloads\mountain.jpg"
pil_img = Image.fromarray(img)
pil_img.save(path)
degrees = 90                                #  For Vertical Flip..
img = np.array(Image.open(r"C:\Users\navna\Downloads\mountain.jpg"))



fig = plt.figure(figsize=(10, 10))
fig.add_subplot(1, 2, 1)
plt.imshow(img)
plt.title("original")
img0 = img.copy()
for _ in range(degrees // 90):
    img0 = img0.transpose(1, 0, 2)
    for j in range(0, img0.shape[1] // 2):
        c = img0[:, j, :].copy()
        img0[:, j, :] = img0[: , img0.shape[1]-j-1, :]
        img0[: , img0.shape[1]-j-1, :] = c
fig.add_subplot(1, 2, 2)
plt.imshow(img0)
plt.title("rotated")
```

```python
img = np.array(Image.open(r"C:\Users\navna\Downloads\mountain.jpg"))
fig = plt.figure(figsize=(10, 10))
fig.add_subplot(1, 2, 1)              # Horizontal flip.....
plt.imshow(np.flipud(img)) 
fig.add_subplot(1, 2, 2)
plt.imshow(np.fliplr(img))
```

```python
# invert Color.

img = np.array(Image.open(r"C:\Users\navna\Downloads\mountain.jpg"))        
img_0 = (img // 64) * 64
img_1 = (img // 128) * 128
img_all = np.concatenate((img, img_0, img_1), axis=1)
plt.figure(figsize=(15, 15))
plt.imshow(img_all)
```

```python
# Grayscale and Black & white image................................

img = np.array(Image.open(r"C:\Users\navna\Downloads\mountain.jpg"))
img_grey = img.sum(2) / (255*3)
img0 = img_grey.copy()
img0 = np.pad(img0, ((100,100),(100,100)), mode='constant')
plt.imshow(img0)

fig = plt.figure(figsize=(10, 10))
img_grey = 255*3 - img_grey        
fig.add_subplot(1, 2, 1)
plt.imshow(img_grey)
plt.title('Grey image')
img = 255 - img
fig.add_subplot(1, 2, 2)
plt.imshow(img)
plt.title('Black And White image')
```
